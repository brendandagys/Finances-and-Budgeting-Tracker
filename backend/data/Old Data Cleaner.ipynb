```python
import pandas as pd

df_purchases = pd.read_json('Purchase-2021-04-15.json')

df_purchases['timestamp'] = df_purchases['date'].astype(str) + ' ' + df_purchases['time']
df_purchases['userName'] = 'Brendan Dagys'

df_purchases.drop(['date', 'time', 'id', 'user', 'category_2', 'amount_2', 'currency', 'exchange_rate'], axis=1, inplace=True)

df_purchases.columns = ['item', 'category_id', 'amount', 'description', 'account_id', 'receiptUrl', 'timestamp', 'userName']

# df_purchases.to_json('Purchases CLEAN.json', orient='records')
df_purchases
```

```python
df_accounts = pd.read_json('AccountUpdate-2021-04-15.json')

df_accounts['userName'] = 'Brendan Dagys'

df_accounts.sort_values('timestamp', inplace=True)

df_accounts['date'] = df_accounts['timestamp'].apply(lambda x: str(x)[:10])

df_accounts.drop_duplicates(keep='last', subset=['account', 'date'], inplace=True)

df_accounts.drop(['id', 'exchange_rate', 'purchase', 'timestamp'], axis=1, inplace=True)

df_accounts.columns = ['account_id', 'value', 'userName', 'timestamp']

# df_accounts.to_json('Account Updates CLEAN.json', orient='records')
df_accounts
```
